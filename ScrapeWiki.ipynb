```python
import re

def ext_remover(text):
    pattern = r'(== جستارهای وابسته ==|== جامعه و فرهنگ ==|== منابع ==|== پیوند به بیرون ==|== پانویس ==|=== سینما ===|== یادداشت‌ها ==|== جدول ==)'
    match = re.search(pattern, text)
    if match:
        return text[:match.start()]
    return text
```

```python
from langchain.schema import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
import re

def chunk_persian_text(text, min_size=200, max_size=1000):
    # Split text into sections and subsections
    sections = re.split(r'(=+)\s*(.*?)\s*\1', text)
    
    chunks = []
    current_chunk = ""
    current_title = "Overview"  
    current_level = 0

    # Process the first paragraph 
    if sections[0].strip():
        chunks.append(Document(page_content=sections[0].strip(), metadata={"title": current_title}))

  
    for i in range(1, len(sections), 3):
        level = len(sections[i].strip())  
        title = sections[i+1].strip()
        content = sections[i+2].strip()

      
        if len(content) < min_size and current_chunk:
            
            current_chunk += f"\n\n{content}"
        else:
            
            if current_chunk:
                chunks.append(Document(page_content=current_chunk, metadata={"title": current_title}))
            
            current_chunk = content
            current_title = title

        if len(current_chunk) > max_size:
            text_splitter = RecursiveCharacterTextSplitter(
                chunk_size=max_size,
                chunk_overlap=200,
                separators=["\n\n", "۔", "،", "\n", " ", ""]  
            )
            sub_chunks = text_splitter.split_text(current_chunk)
            for sub in sub_chunks:
                chunks.append(Document(page_content=sub, metadata={"title": current_title}))
            current_chunk = ""

    # Add the last remaining chunk
    if current_chunk:
        chunks.append(Document(page_content=current_chunk, metadata={"title": current_title}))

    return chunks
```

```python
import os

chunks = []
folder_path = "Documnets"  
file_list = os.listdir(folder_path)

for f in file_list:
    file_path = os.path.join(folder_path, f)
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        chunks.extend(chunk_persian_text(text))
```

```python
# Removing duplicates for the embedding :

vectorstore = FAISS.load_local("app/faiss_index", embedding_model, allow_dangerous_deserialization=True)

docs = vectorstore.docstore._dict
ids_to_remove = set()
seen_contents = set()

# Identify duplicates
for doc_id, doc in docs.items():
    content = doc.page_content
    if content in seen_contents:
        ids_to_remove.add(doc_id)
    else:
        seen_contents.add(content)

# Remove duplicates from the index
vectorstore.delete(ids_to_remove)
vectorstore.save_local("faiss_index_deduplicated")
```
